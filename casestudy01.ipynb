# **Case Study on Testing of Hypothesis**

A company started to invest in digital marketing as a new way of their product
promotions. For that they collected data and decided to carry out a study on it.
● The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.
● The company needs to check whether there is any dependency between
the features “Region” and “Manager”.


```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#loading sales dataset
df=pd.read_csv('/content/Sales_add (1).csv')
df
```

Question No. 1
The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.

```python
#import statistics as stat
from scipy.stats import ttest_ind
```

Hypothesis
Null hypothesis     : There is no increase in sales after digital marketing"
Alternate Hypothesis: There is an increase in sales after digital marketing"

```python
#computing the p value and statistic
t_score,p_value = ttest_ind(df.iloc[:,3], df.iloc[:,4])
```

```python
# checking weather to accept or reject hypothesis
if p_value < 0.05:
  print('Reject Null Hypothesis')
  print('There is an increase in sales after digital marketing.')
else:
  print('Accept Null Hypothesis')
  print('There is no increase in sales after digital marketing.')
```

Question no.2
The company needs to check whether there is any dependency between
thefeatures “Region” and “Manager”
Null Hypothesis     : There is no dependency between Manager and Region
Alternate Hypothesis: There is a dependece between Manager and Region

```python
pd.crosstab(df['Manager'], df['Region'])
```

```python
#importing library
from scipy.stats import chi2_contingency
```

```python
#creating a pandas crosstab with Manager and Region column of data
manager_x_region = pd.crosstab(df['Manager'], df['Region'])
```

```python
result = chi2_contingency(manager_x_region)
print(result)
print("")
if result.pvalue < 0.05:
  print('Reject Null Hypothesis')
  print('There is a dependency between Manager and Region.')
else:
  print('Accept Null Hypothesis')
  print('There is no dependence between Manager and Region')
```
