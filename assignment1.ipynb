5. Perform meaningful visualizations using the dataset. Bring at least 3

```python
#using scatter plot
plt.scatter(data['petal_width'],data['sepal_length'],color='Orange',edgecolor='red')
plt.title('Petal Width v/s Sepal Length')
plt.xlabel('Petal Width')
plt.ylabel('Sepal Length')
plt.show()
```
