Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in ['Gender', 'Dependents','Self_Employed','Married','Loan_ID','Education','Property_Area', 'Loan_Status']:
  train1[i] = encoder.fit_transform(train1[i])
for i in ['Gender', 'Dependents','Self_Employed','Married','Loan_ID','Education','Property_Area']:
  test1[i] = encoder.fit_transform(test1[i])
```

```python
#Missing value handling of test data
freqgraph = train1.select_dtypes(include = 'float64')
freqgraph.hist(figsize = (20,15))
```

```python
train1.skew()
```

```python
test1.skew()
```

```python
for i in ['Gender', 'Dependents','Self_Employed']:
    train1[i] = train1[i].fillna(train1[i].mode())
    test1[i] = test1[i].fillna(test1[i].mode())
```

```python
for i in ['LoanAmount', 'Loan_Amount_Term','Credit_History']:
    train1[i] = train1[i].fillna(train1[i].median())
    test1[i] = test1[i].fillna(test1[i].median())
```

Random Forest

```python
def build_rf_model(a,b):
    from sklearn.ensemble import RandomForestClassifier

    model = RandomForestClassifier()
    model.fit(a,b)
    return model
def build_lg_model(a,b):
    from sklearn.linear_model import LogisticRegressionCV

    model = LogisticRegressionCV()
    model.fit(a,b)
    return model
```

```python
model_rf = build_rf_model(X,y)
```

Prediction for test data

```python

test1['Loan_Status'] = model_rf.predict(test1)
```
